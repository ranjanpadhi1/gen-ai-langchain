## Ollama Embeddings

https://github.com/ollama/ollama

```python
from langchain_community.embeddings import OllamaEmbeddings

embeddings = OllamaEmbeddings(model="gemma2:2b") ## By default it uses Llama 2 from Meta
```

#### Create Vector Embeddings using Gemma2:2b model

```python
docs = ["Alpha is the first letter of greek alphabet"]
vector = embeddings.embed_documents(docs)
```
